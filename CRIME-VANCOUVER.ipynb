```python
#CLEANING THE DATA
```

```python
crimev.isnull().sum()
```

```python
crimev['HOUR']=crimev['HOUR'].fillna(method='bfill')
crimev['MINUTE']=crimev['MINUTE'].fillna(method='bfill')
crimev['HUNDRED_BLOCK']=crimev['HUNDRED_BLOCK'].fillna(method='bfill')
crimev['NEIGHBOURHOOD']=crimev['NEIGHBOURHOOD'].fillna(method='bfill')
```

```python
crimev.isnull().sum()
```

```python
crimev['HOUR']=crimev['HOUR'].fillna(method='ffill')
crimev['MINUTE']=crimev['MINUTE'].fillna(method='ffill')
crimev['HUNDRED_BLOCK']=crimev['HUNDRED_BLOCK'].fillna(method='ffill')
crimev['NEIGHBOURHOOD']=crimev['NEIGHBOURHOOD'].fillna(method='ffill')
```

```python
#VISUALIZING THE DATA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
crr = crimev.corr()
```

```python
sns.heatmap(crr,xticklabels=crr.columns,yticklabels=crr.columns,annot=True)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3,3,1)
sns.distplot(crimev['MONTH'],color='blue')
plt.title('Distribution of crimes in vancouver per month')
plt.figure(figsize=(15,15))
plt.subplot(3,3,2)
sns.distplot(crimev['DAY'],color='red')
plt.title('Distribution of crimes in vancouver per day')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3,3,1)
sns.boxplot(x='TYPE',y='MONTH',data=crimev)
plt.title('Crimes in vancouver per month across year box plot')


plt.figure(figsize=(15,15))
plt.subplot(3,3,2)
sns.boxplot(x='YEAR',y='DAY',data=crimev)
plt.title('Crimes in vancouver per day across year box plot')

plt.show()
```

```python
from matplotlib import style
plt.figure(figsize=(15,15))
plt.subplot(3,3,1)
crimev.groupby('YEAR')['MONTH'].mean().plot(kind='bar')
plt.title("Average crimes in vancouver per month across each year")

plt.figure(figsize=(15,15))
plt.subplot(3,3,2)
crimev.groupby('YEAR')['DAY'].mean().plot(kind='bar')
plt.title("Average crimes in vancouver per day across each year")
plt.tight_layout()
style.use('classic')
plt.show()
```
